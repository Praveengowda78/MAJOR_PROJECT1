```python
#Take the data and create dataframe

import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Salary_Data.csv")
df
```

```python
x=df.iloc[:,0:1].values #  .values gives 1 dimensional
x
```

```python
y=df['Salary'].values
y
```

```python
#5. Train and test variables- If we have more data, preform step no. 5 better accuracy
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
# the variables names can change, but the variable order cannot change
```

```python
#7. Run a classifier, regressor, cluster(applying a suitable algorith)
#sklearn.linear_model- package(collection of libraries) ,LogisticRegression-library name
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_train,y_train)
```

```python
#9. Predict the o/p
y_pred=model1.predict(x_test)#using the i/p testing values , we predict the o/p
y_pred#predicted o/p values
```

```python
y_test #actual o/p values
```

```python
#10. accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```python
model1.predict([[3.2	]])
```

```python
#7. Run a classifier, regressor, cluster(applying a suitable algorith)
#sklearn.linear_model- package(collection of libraries) ,LogisticRegression-library name
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#8. Fit the model(mapping or plotting the inputs with the output)
#LinearRegression.fit(x,y)
model.fit(x,y) #We are plotting the values of x and y inside the LinearRegression library
```

```python
#9. Predict the output
y_pred=model.predict(x) # using the input values we predict the output
y_pred
```
